```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw/globalmech_data.csv", parse_dates=["Date"], index_col="Date")

df.plot(subplots=True, figsize=(12, 10))
plt.show()
```

What we Observe:
* Trend in revenue streams
* Seasonality (monthly patterns)
* Correlation with:
    * Install Base
    * USD
