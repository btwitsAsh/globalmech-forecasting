```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Create date range (5 years monthly)
dates = pd.date_range(start='2019-01-01', end='2024-12-01', freq='MS')
n = len(dates)

# Base trends
trend = np.linspace(100, 200, n)
seasonality = 10 * np.sin(np.arange(n) * 2 * np.pi / 12)

# External factors
usd_inr = 70 + np.cumsum(np.random.normal(0.2, 0.5, n))
shipping_index = 50 + np.random.normal(0, 5, n)
install_base = 500 + np.cumsum(np.random.randint(5, 15, n))

# Revenue streams
dom_cap = trend * 1.5 + seasonality + np.random.normal(0, 10, n)
dom_spares = install_base * 0.3 + np.random.normal(0, 20, n)

exp_cap = trend * 1.2 + shipping_index * (-0.5) + np.random.normal(0, 10, n)
exp_spares = install_base * 0.25 + usd_inr * 2 + np.random.normal(0, 15, n)

# Create DataFrame
df = pd.DataFrame({
    "Date": dates,
    "Dom_Cap": dom_cap,
    "Dom_Spares": dom_spares,
    "Exp_Cap": exp_cap,
    "Exp_Spares": exp_spares,
    "Install_Base": install_base,
    "USD_INR": usd_inr,
    "Shipping_Index": shipping_index
})

df.set_index("Date", inplace=True)

df.to_csv("../data/raw/globalmech_data.csv")

df.head()
```

* Simulated trend + seasonality + noise
* Created external variables (multivariate drivers)
* Modeled realistic business relationships:
* Spares depend on install base
* Export depends on USD
* Capital depends on trend + delays
